```python
# Faça um Programa que peça dois números e imprima a soma.

numero1 = int(input("Digite um número: "))

numero2 = int(input("Digite mais um número: "))

soma = numero1 + numero2

print("A soma dos números", numero1, "e", numero2, "é", soma)
```

```python
# Faça um Programa que peça 3 notas e mostre a média.

nota1 = float(input("Digite a primeira nota: "))
nota2 = float(input("Digite a segunda nota: "))
nota3 = float(input("Digite a terceira nota: "))

media = (nota1 + nota2 + nota3) / 3

print("A média é", media)
```

```python
# Faça um Programa que pergunte quanto você ganha por hora e o número de horas trabalhadas nos mês. Calcule e mostre o total do seu salário no referido mês.

ganhoHora= float(input("Digite quanto você ganha por hora: "))
horasTrabalhadas = float(input("Digite quantas horas você trabalhou no mês: "))

salario = ganhoHora * horasTrabalhadas

print("O seu salário do mês será de R$", salario)
```

```python
# João é pecador no litoral do Recife. Toda vez que João traz um peso de peixes maior que o estabelecido pelo regulamento de pesca do estado de Pernambuco (50 quilos) deve pagar uma multa de R$ 4,00 por quilo excedente.

# Facilite a vida de João! Faça um programa que leia a variável peso (peso dos peixes) e calcule o excesso.

# Gravar na variável excesso a quantidade de quilos além do limite e na variável multa o valor da multa que João deverá pagar. Imprima os dados do programa com as mensagens para alertar João.

print("Só informe o peso dos peixes se o mesmo exceder 50 quilos!")

peso = float(input("Informe o peso dos peixes: "))

excesso = peso - 50
multa = 4 * excesso

print("Seu excesso foi de", excesso)
print("Sua multa será de R$", multa)
```
