```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Pokemon.csv')
```

1. Quantos tipos de pokemons existem?
2. Quais são os tipos?
3. Como está distribuída a quantidade de cada tipo? Inclua um gráfico.
4. Qual é o pokemon com maior nível de HP?
5. Qual é o pokemon com menor nível de HP?



```python
# 1. Quantos tipos de pokemons existem?

# Conta os tipos de pokemons.
tipo_counts = df['Type 1'].value_counts()

# Cria o gráfico de barras.
tipo_counts.plot(kind='bar', figsize=(10, 6), color='skyblue')

# Configuraçõe do gráfico
plt.title('Número de Pokemons por Tipo')
plt.xlabel('Tipo')
plt.ylabel('Número de Pokemons')
plt.xticks(rotation=45)  # Rotaciona os rótulos do eixo x para facilitar a leitura.

# Mostra o gráfico.
plt.tight_layout()
plt.show()

print("Número de tipos de pokemons:", len(tipo_counts))
```

```python
# 2. Quais são os tipos?

tipo_counts.plot(kind='bar', figsize=(10, 6), stacked=True)

plt.title('Distribuição de Tipos de Pokemons')
plt.xlabel('Tipo')
plt.ylabel('Número de Pokemons')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()


```

```python
# 3. Como está distribuída a quantidade de cada tipo? Inclua um gráfico.

tipo_counts = tipo_counts.sort_values(ascending=False)

plt.figure(figsize=(12, 6))
tipo_counts.plot(kind='bar', color='skyblue')

plt.title('Distribuição da Quantidade de Cada Tipo de Pokémon')
plt.xlabel('Tipo')
plt.ylabel('Número de Pokémons')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# 4. Qual é o pokemon com maior nível de HP?

# Encontra o índice do Pokémon com o maior nível de HP.
indice_max_hp = df['HP'].idxmax()

# Extrai informações do Pokémon com maior HP.
pokemon_max_hp = df.loc[indice_max_hp]

print("Pokemon com maior nível de HP:")
print(pokemon_max_hp[['Name', 'HP']])
```

```python
# 5. Qual é o pokemon com menor nível de HP?

indice_min_hp = df['HP'].idxmin()

pokemon_min_hp = df.loc[indice_min_hp]

print("Pokemon com menor nível de HP:")
print(pokemon_min_hp[['Name', 'HP']])
```
